```python
# Entanglement
import qiskit as qs
import math
import numpy as np
```

```python
statevector_simulator = qs.Aer.get_backend("statevector_simulator")
qasm_simulator = qs.Aer.get_backend("qasm_simulator")
unitary_simulator = qs.Aer.get_backend("unitary_simulator")
```

```python
# Controlled X gate
q_circuit_1 = qs.QuantumCircuit(2)
q_circuit_1.cx(0, 1)
q_circuit_1.draw()
```

```python
result_1 = qs.execute(q_circuit_1, unitary_simulator).result()
result_2 = qs.execute(q_circuit_1, statevector_simulator).result()
print("Unitary Matrix: ", result_1.get_unitary())
print("Statevector: ", result_2.get_statevector())
qs.visualization.plot_histogram(result_2.get_counts())
```

```python
q_circuit_2 = qs.QuantumCircuit(2)
q_circuit_2.h(0)
q_circuit_2.draw()
```

```python
result_3 = qs.execute(q_circuit_2, unitary_simulator).result()
result_4 = qs.execute(q_circuit_2, statevector_simulator).result()
print("Unitary Matrix: ", result_3.get_unitary())
print("Statevector: ", result_4.get_statevector())
qs.visualization.plot_histogram(result_4.get_counts())
```

```python
q_circuit_3 = qs.QuantumCircuit(2)
q_circuit_3.h(0)
q_circuit_3.cx(0, 1)
q_circuit_3.draw()
```

```python
result_5 = qs.execute(q_circuit_3, unitary_simulator).result()
result_6 = qs.execute(q_circuit_3, statevector_simulator).result()
print("Unitary Matrix: ", result_5.get_unitary())
print("Statevector: ", result_6.get_statevector())
qs.visualization.plot_histogram(result_6.get_counts())
# This shows that if qubit 0 is measured as 0, then qubit 1 has to be 0
# If qubit 0 is 1 then qubit 1 must be 1.
# These two qubits are now entangled! This particular state is called the Bell State.
# It’s important to note that the measurement result is random, and the measurement statistics of one qubit are not affected by any operation on the other qubit. Because of this, there is no way to use shared quantum states to communicate. This is known as the no-communication theorem.[1]
```

```python
q_circuit_4 = qs.QuantumCircuit(2)
q_circuit_4.h(0)
q_circuit_4.x(1)
q_circuit_4.cx(0, 1)
q_circuit_4.draw()
```

```python
result_7 = qs.execute(q_circuit_4, unitary_simulator).result()
result_8 = qs.execute(q_circuit_4, statevector_simulator).result()
print("Unitary Matrix: ", result_7.get_unitary())
print("Statevector: ", result_8.get_statevector())
qs.visualization.plot_histogram(result_8.get_counts())
# here qubits are entangled but oppositely linked
```
