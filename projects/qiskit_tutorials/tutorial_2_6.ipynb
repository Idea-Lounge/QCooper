```python
# Entanglement
import qiskit as qs
```

```python
q_circuit_1 = qs.QuantumCircuit(2)
q_circuit_1.x(1)
q_circuit_1.h(0)
q_circuit_1.cx(0, 1)
q_circuit_1.draw()
```

```python
q_simulator = qs.Aer.get_backend("statevector_simulator")
```

```python
result_1 = qs.execute(q_circuit_1, q_simulator).result()
```

```python
result_1.get_statevector() # Bell State! if we detect qubit 0, we know the value qubit 1 will have
```

```python
qs.visualization.plot_histogram(result_1.get_counts())
```
